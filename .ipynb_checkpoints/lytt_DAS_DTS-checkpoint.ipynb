```python
# Function to plot training process
def plot_loss(fit):
    epochs     = len(fit.history['loss'])
    iterations = np.arange(epochs)
    plt.plot(iterations, fit.history['loss'],     '-', label='loss')
    plt.plot(iterations, fit.history['val_loss'], '-', label='validation loss')
    plt.title('Training: Loss vs epochs'); plt.legend()
    plt.xlabel('Epochs'); plt.ylabel('Loss')
    plt.xticks(iterations[::epochs//10])
```

```python
from keras.layers import MaxPooling2D, UpSampling2D, RepeatVector, Conv2D, ZeroPadding1D, Conv2DTranspose
from tensorflow import expand_dims
from tensorflow.keras.layers import concatenate
```

```python
X_das = np.expand_dims(das, -1)
X_dts = np.expand_dims(dts, -1)
y_flow = flow_norm

print('DAS norm: ', X_das.shape)
print('DTS norm: ', X_dts.shape)
print('Flow norm:', y_flow.shape)
```

```python
def das_Unet(act=LeakyReLU(alpha=0.3)):
    tf.keras.backend.clear_session()

    image = tf.keras.Input((X_das.shape[1], X_das.shape[-1]),name='input')

    conv1 = Conv1D(16, 3, activation=act, padding='same')(image)
    conv1 = Conv1D(16, 3, activation=act, padding='same')(conv1)
    conv1 = BatchNormalization()(conv1)

    pool1 = MaxPooling1D(pool_size=2)(conv1)
    conv2 = Conv1D(32, 3, activation=act, padding='same')(pool1)
    conv2 = Conv1D(32, 3, activation=act, padding='same')(conv2)
    conv2 = BatchNormalization()(conv2)

    pool2 = MaxPooling1D(pool_size=2)(conv2)
    conv3 = Conv1D(64, 3, activation=act, padding='same')(pool2)
    conv3 = Conv1D(64, 3, activation=act, padding='same')(conv3)
    conv3 = BatchNormalization()(conv3)

    pool3 = MaxPooling1D(pool_size=2)(conv3)
    conv4 = Conv1D(128, 3, activation=act, padding='same')(pool3)
    conv4 = Conv1D(128, 3, activation=act, padding='same')(conv4)
    conv4 = BatchNormalization()(conv4)
    latent = conv4

    up7 = Conv1DTranspose(64, 3, strides=2, padding='same')(conv4)
    up7 = concatenate([up7, conv3])
    conv7 = Conv1D(64, 3, activation=act, padding='same')(up7)
    conv7 = Conv1D(64, 3, activation=act, padding='same')(conv7)
    conv7 = BatchNormalization()(conv7)

    up8 = Conv1DTranspose(32, 3, strides=2, padding='same')(conv7)
    up8 = concatenate([up8, conv2])
    conv8 = Conv1D(32, 3, activation=act, padding='same')(up8)
    conv8 = Conv1D(32, 3, activation=act, padding='same')(conv8)
    conv8 = BatchNormalization()(conv8)

    up9 = Conv1DTranspose(16, 3, strides=2, padding='same')(conv8)
    up9 = concatenate([up9, conv1])
    conv9 = Conv1D(16, 3, activation=act, padding='same')(up9)
    conv9 = Conv1D(16, 3, activation=act, padding='same')(conv9)
    conv9 = BatchNormalization()(conv9)

    out = Conv1D(1, 1, activation='linear')(conv9)
    
    das_m2m = Model(inputs=[image], outputs=[out])
    das_m2z = Model(inputs=[image], outputs=[latent])

    return das_m2m, das_m2z
```

```python
das_m2m, das_m2z = das_Unet()
print('AutoEncoder # of parameters: {:,}'.format(das_m2m.count_params()))
print('AutoEncoder # of parameters: {:,}'.format(das_m2z.count_params()))
```

```python
das_m2m.compile(optimizer = tf.keras.optimizers.Nadam(learning_rate=1e-3),
                    loss    = 'mse',
                    metrics = ['mse'])

fit_das = das_m2m.fit(X_das, X_das,
                        shuffle          = False,
                        epochs           = 20,
                        batch_size       = 128,
                        validation_split = 0.2,
                        verbose          = 0)
plot_loss(fit_das)
```

```python
das_pred = das_m2m.predict(X_das).squeeze().astype('float64')
print(das_pred.shape)

das_z = das_m2z.predict(X_das).squeeze().astype('float64')
print(das_z.shape)
```

```python
def dts_Unet(act=LeakyReLU(alpha=0.3)):
    tf.keras.backend.clear_session()

    image = tf.keras.Input((X_dts.shape[1], X_dts.shape[-1]),name='input')

    conv1 = Conv1D(16, 3, activation=act, padding='same')(image)
    conv1 = Conv1D(16, 3, activation=act, padding='same')(conv1)
    conv1 = BatchNormalization()(conv1)

    pool1 = MaxPooling1D(pool_size=2)(conv1)
    conv2 = Conv1D(32, 3, activation=act, padding='same')(pool1)
    conv2 = Conv1D(32, 3, activation=act, padding='same')(conv2)
    conv2 = BatchNormalization()(conv2)

    pool2 = MaxPooling1D(pool_size=2)(conv2)
    conv3 = Conv1D(64, 3, activation=act, padding='same')(pool2)
    conv3 = Conv1D(64, 3, activation=act, padding='same')(conv3)
    conv3 = BatchNormalization()(conv3)

    pool3 = MaxPooling1D(pool_size=2)(conv3)
    conv4 = Conv1D(128, 3, activation=act, padding='same')(pool3)
    conv4 = Conv1D(128, 3, activation=act, padding='same')(conv4)
    conv4 = BatchNormalization()(conv4)
    latent = conv4

    up7 = Conv1DTranspose(64, 3, strides=2, padding='same')(conv4)
    up7 = concatenate([up7, conv3])
    conv7 = Conv1D(64, 3, activation=act, padding='same')(up7)
    conv7 = Conv1D(64, 3, activation=act, padding='same')(conv7)
    conv7 = BatchNormalization()(conv7)

    up8 = Conv1DTranspose(32, 3, strides=2, padding='same')(conv7)
    up8 = concatenate([up8, conv2])
    conv8 = Conv1D(32, 3, activation=act, padding='same')(up8)
    conv8 = Conv1D(32, 3, activation=act, padding='same')(conv8)
    conv8 = BatchNormalization()(conv8)

    up9 = Conv1DTranspose(16, 3, strides=2, padding='same')(conv8)
    up9 = concatenate([up9, conv1])
    conv9 = Conv1D(16, 3, activation=act, padding='same')(up9)
    conv9 = Conv1D(16, 3, activation=act, padding='same')(conv9)
    conv9 = BatchNormalization()(conv9)

    out = Conv1D(1, 1, activation='linear')(conv9)
    
    dts_m2m = Model(inputs=[image], outputs=[out])
    dts_m2z = Model(inputs=[image], outputs=[latent])

    return dts_m2m, dts_m2z
```

```python
dts_m2m, dts_m2z = dts_Unet()
print('AutoEncoder # of parameters: {:,}'.format(dts_m2m.count_params()))
print('AutoEncoder # of parameters: {:,}'.format(dts_m2z.count_params()))

dts_m2m.compile(optimizer = tf.keras.optimizers.Nadam(learning_rate=1e-2),
                    loss    = 'mse',
                    metrics = ['mse'])

fit_dts = dts_m2m.fit(X_dts, X_dts,
                        shuffle          = False,
                        epochs           = 100,
                        batch_size       = 24,
                        validation_split = 0.2,
                        verbose          = 0)
plot_loss(fit_dts)
```

```python
def conv_up_block(filt, kern, pool, act, inp):
    x = Conv1D(filters=filt, kernel_size=kern, padding='same', activation=act)(inp)
    x = Conv1D(filters=filt, kernel_size=kern, padding='same', activation=act)(x)
    x = BatchNormalization()(x)
    x = MaxPooling1D(pool_size=pool)(x)
    return x

def conv_down_block(filt, kern, pool, act, inp):
    x = Conv1D(filters=filt, kernel_size=kern, padding='same', activation=act)(inp)
    x = Conv1D(filters=filt, kernel_size=kern, padding='same', activation=act)(x)
    x = BatchNormalization()(x)
    x = UpSampling1D(size=pool)(x)
    return x

def make_flow_net(act=LeakyReLU(alpha=0.2), kern=3):
    keras.backend.clear_session()
    inp = Input(shape=(z_dual.shape[1], z_dual.shape[-1]))
    x = conv_down_block(filt=96, kern=kern, pool=2, act=act, inp=inp)
    x = conv_down_block(filt=64, kern=kern, pool=2, act=act, inp=x)
    x = conv_down_block(filt=32, kern=kern, pool=2, act=act, inp=x)
    x = conv_down_block(filt=16, kern=kern, pool=2, act=act, inp=x)
    x = conv_up_block(filt=16, kern=kern, pool=3, act=act, inp=x)
    x = conv_up_block(filt=32, kern=kern, pool=3, act=act, inp=x)
    x = conv_up_block(filt=64, kern=kern, pool=3, act=act, inp=x)
    x = conv_up_block(filt=128, kern=kern, pool=3, act=act, inp=x)
    x = conv_up_block(filt=200, kern=kern, pool=3, act=act, inp=x)
    out = tf.squeeze(x, axis=1)
    return Model(inp, out)
```

```python
flow_net = make_flow_net()
print('AutoEncoder # of parameters: {:,}'.format(flow_net.count_params()))
#flow_net.summary()
```

```python
flow_net.compile(optimizer = tf.keras.optimizers.Nadam(learning_rate=1e-3),
                    loss    = 'mae',
                    metrics = ['mae'])

fit_flow = flow_net.fit(z_dual, flow_expanded,
                        shuffle          = False,
                        epochs           = 20,
                        batch_size       = 24,
                        validation_split = 0.2,
                        verbose          = 0)
plot_loss(fit_flow)
```

```python
flow_pred = flow_net.predict(z_dual).T
flow_pred.shape
```

```python
flow_expanded_2 = np.tile(flowrates, (flow_pred.shape[-1],1)).T
flow_expanded_2.shape
```

```python
mse  = mean_squared_error(flow_expanded_2, flow_pred)
ssim = structural_similarity(flow_expanded_2, flow_pred)
print('MSE={:.3e} | SSIM={:.5f}'.format(mse,ssim))
```
